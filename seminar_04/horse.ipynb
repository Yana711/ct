```python
import heapq

N  = int(input('Введите размер доски:'))
a = list(map(int, input('Введите начальное положение(x_1, y_1) c учетом, что отсчет происходит с нуля:').split()))
b = list(map(int, input('Введите начальное положение(x_2, y_2) c учетом, что отсчет происходит с нуля:').split()))
x_1 = a[0]
y_1 = a[1]
x_2 = b[0]
y_2 = b[1]

start = x_1 * N + y_1
stop = x_2 * N + y_2

vert_num = N**2
BIG_NUM = 10000000
```

```python
def moves(x_1, y_1):
    coordinates = []

    if x_1 + 2<=N-1 and x_1 + 2>=0 and y_1 + 1<=N-1 and y_1 + 1 >=0:
        coordinates.append([x_1 + 2, y_1 + 1])
        
    if x_1 + 2<=N-1 and x_1 + 2>=0 and y_1 - 1<=N-1 and y_1 - 1 >=0:    
        coordinates.append([x_1 + 2, y_1 - 1])
        
    if x_1 - 2<=N-1 and x_1 - 2>=0 and y_1 + 1<=N-1 and y_1 + 1 >=0:
        coordinates.append([x_1 - 2, y_1 + 1])
        
    if x_1 - 2<=N-1 and x_1 - 2>=0 and y_1 - 1<=N-1 and y_1 - 1 >=0:    
        coordinates.append([x_1 - 2, y_1 - 1])
        
    if x_1 + 1<=N-1 and x_1 + 1>=0 and y_1 + 2<=N-1 and y_1 + 2 >=0:    
        coordinates.append([x_1 + 1, y_1 + 2])
       
    if x_1 + 1<=N-1 and x_1 + 1>=0 and y_1 - 2<=N-1 and y_1 - 2 >=0:    
        coordinates.append([x_1 + 1, y_1 - 2])
        
    if x_1 - 1 <=N-1 and x_1 - 1>=0 and y_1 + 2<=N-1 and y_1 + 2 >=0:    
        coordinates.append([x_1 - 1, y_1 + 2])
        
    if x_1 - 1<=N-1 and x_1 - 1 >=0 and y_1 - 2<=N-1 and y_1 - 2 >=0:    
        coordinates.append([x_1 - 1, y_1 - 2])
        
    
    
    return coordinates
```

```python
adj_list = []

st = []
for i in range (0, N):
    for y in range (0, N):
        step = moves(i, y)
        st = []
        for s in range(0, len(step)):
            f = step[s]
            
            x_step = f[0]
            y_step = f[1]
            
            st.append(x_step * N + y_step)
        if len(set(st)) != len(st): 
            st = list(set(st))
        adj_list.append(st) 
print(adj_list)
print("start:", start)
print("stop:", stop)
```

```python
edge_list = []


for i in range(0, len(adj_list)):
    k = adj_list[i]
    ed = []
    for y in range(0, len(k)):
        edge_list.append([i, k[y]])

print(edge_list)  
```

```python
weight_list = [1] * len(edge_list)
print(weight_list)
```

```python
def construct_adj_matrix(edge_list, weight_list, N):
    adj_matrix = [[BIG_NUM for _ in range(N**2)] for _ in range(N**2)]
    
    for edge, weight in zip(edge_list, weight_list):
        src = edge[0]
        dest = edge[1]
        adj_matrix[src][dest] = weight
        
    return adj_matrix   

adj_matrix = construct_adj_matrix(edge_list, weight_list, N)
adj_matrix 
```

```python
def dijkstra(adj_list, adj_matrix, src):
    vert_num = len(adj_list)
    dist = [BIG_NUM for _ in range(vert_num)]
    used = [False for _ in range(vert_num)]
    dist[src] = 0
    heap = [(0, src)] 

    while not all(used):
        heap_2 = heapq.heappop(heap)
        u = list(heap_2)[1]
        if used[u]:
            continue
        used[u] = True 
        
        for v in adj_list[u]:
                weight = adj_matrix[u][v]
                if dist[v] > dist[u] + weight:
                    dist[v] = dist[u] + weight  
                    heapq.heappush(heap, (dist[v], v))
    return dist
```

```python
mov = dijkstra(adj_list, adj_matrix, start)
res = mov[stop]
print('Минимальное количество шагов:', res)
```
